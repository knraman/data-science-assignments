```python
X = data.loc[:, data.columns !="size_category"]
Y = data["size_category"]
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state =0)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
x_train
```

```python
from sklearn import svm 
from sklearn.svm import SVC
```

```python
model = SVC(kernel='linear', C=0.01)
```

```python
model
```

```python
model.fit(x_train,y_train)
```

```python
pred = model.predict(x_test)
```

```python
pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(pred , y_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(pred,y_test)
```

```python
sns.heatmap(cm , annot = True , center = 0 )
```

### grid search to get better accuracy scorw

```python
from sklearn.model_selection import GridSearchCV 
```

```python
parameters = {'C': [0.001, 0.005, 0.01, 0.05, 0.1, 0.5, 1, 5, 10, 50, 100, 500, 1000, 5000],              
 'gamma':[0.001, 0.01, 0.1, 0.5, 1],           
    'kernel': ['rbf','poly','linear']}
```

```python
model = SVC()
```

```python
gcv = GridSearchCV(model,parameters,cv=10)
```

```python
gcv.fit(x_train,y_train)
```

```python
gcv.best_score_
```

```python
gcv.best_params_
```

### final model

```python
final_model = SVC(kernel='linear', C=10,gamma = 0.001)
```

```python
final_model
```

```python
final_model.fit(x_train,y_train)
```

```python
pred = final_model.predict(x_test)
```
