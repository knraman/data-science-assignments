# SVM model

```python
from sklearn.svm import SVC
```

```python
from sklearn.metrics import accuracy_score
```

```python
model = SVC()
```

```python
parameters = {'C': [0.001, 0.005, 0.01, 0.05, 0.1, 0.5, 1, 5, 10, 50, 100, 500, 1000, 5000],                  
 'gamma':[0.001, 0.01, 0.1, 0.5, 1],           
    'kernel': ['rbf','poly','linear'] } 
```

```python
c_range = [0.001, 0.005, 0.01, 0.05, 0.1, 0.5, 1, 5, 10, 50, 100]
gamma_range = [0.001, 0.01, 0.1, 0.5, 1]

for i in c_range:
    for j in gamma_range:
        model = SVC(C = i ,gamma = j , kernel = 'linear')
        model.fit(X_train , y_train)
        pred = model.predict(X_test)
        score = accuracy_score(pred,y_test)
        print('score with',i,'and',j,'=',score )
```

```python
final_model = SVC(C = 0.1 ,gamma = 0.01 , kernel = 'linear')
final_model.fit(X_train , y_train)
pred = final_model.predict(X_test)
score = accuracy_score(pred,y_test)
```

```python
score
```

```python
final_model = SVC(C = 0.1 ,gamma = 0.01 , kernel = 'poly')
final_model.fit(X_train , y_train)
pred = final_model.predict(X_test)
score = accuracy_score(pred,y_test)
score
```

```python
final_model = SVC(C = 0.1 ,gamma = 0.01 , kernel = 'rbf')
final_model.fit(X_train , y_train)
pred = final_model.predict(X_test)
score = accuracy_score(pred,y_test)
score
```

```python
final_model = SVC(C = 0.1 ,gamma = 0.01 , kernel = 'sigmoid')
final_model.fit(X_train , y_train)
pred = final_model.predict(X_test)
score = accuracy_score(pred,y_test)
score
```

# using the non linear margin like RBF will give us a better accuracy score . therefore we can use the final model with C as 0.1 , gamma as 0.01 and Kernel as RBF
